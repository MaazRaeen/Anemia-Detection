# 📚 Complete Project Guide: Anemia Detection System

## 🎯 Project Overview
This is a **non-invasive anemia detection system** using deep learning and eye images. The system uses **three pre-trained CNN models** that are fine-tuned on a dataset of 307 eye images, then combined using ensemble learning for improved accuracy.

---

## 📁 Project Files & Their Purpose

### 1️⃣ **VGG.py** - VGG16 Model Training
**What it does:**
- Trains a VGG16 model using 5-fold cross-validation
- VGG16 is a 16-layer deep CNN with small convolutional filters
- Uses transfer learning (pre-trained ImageNet weights)
- Fine-tunes only the top layers on anemia dataset

**Process:**
1. Loads the cleaned dataset (307 images)
2. Splits into 5 folds for cross-validation
3. For each fold:
   - Trains for 30 epochs on training data
   - Fine-tunes for 15 more epochs with lower learning rate
   - Saves best model checkpoint
4. Handles class imbalance by upsampling minority class

**Output Files:**
- `best_vgg16_fold_1.keras` - Best VGG16 model (56M)
- (Would generate fold_2, fold_3, etc. for full 5-fold)

**Performance:** 46.77% test accuracy ⚠️ (Not recommended)

---

### 2️⃣ **DenseNet.py** - DenseNet121 Model Training
**What it does:**
- Trains a DenseNet121 model using 5-fold cross-validation
- DenseNet uses dense connections between layers (each layer connected to all previous)
- More efficient feature reuse than VGG
- Similar training process to VGG

**Process:**
1. Loads dataset and creates 5 folds
2. For each fold:
   - Trains for 30 epochs
   - Fine-tunes for 15 epochs
   - Saves best model
3. Applies data augmentation (rotation, flipping, zooming)

**Output Files:**
- `best_densenet_fold_1.keras` - Best DenseNet model (28M)

**Performance:** 80.65% test accuracy ✅ (Good)

---

### 3️⃣ **Inception.py** - InceptionV3 Model Training
**What it does:**
- Trains an InceptionV3 model using 5-fold cross-validation
- InceptionV3 uses multi-scale parallel convolutions (Inception modules)
- Captures features at multiple scales simultaneously
- Best performer among the three

**Process:**
1. Loads dataset and creates 5 folds
2. For each fold:
   - Trains for 30 epochs
   - Fine-tunes top 2 Inception blocks
   - Saves best model
3. Same augmentation as other models

**Output Files:**
- `best_inception_fold_1.keras` - Best InceptionV3 model (84M)

**Performance:** 88.71% test accuracy 🏆 (BEST!)

---

### 4️⃣ **ensembleIncepDen.py** - 2-Model Ensemble
**What it does:**
- Combines predictions from DenseNet121 + InceptionV3
- Uses soft voting (averages probability predictions)
- More stable than single model
- Good for production if InceptionV3 unavailable

**Process:**
1. Loads both trained models
2. Gets predictions from each on test images
3. Averages the probabilities
4. Takes class with higher average probability
5. Evaluates using confusion matrix & metrics

**Output:**
- Prints classification reports
- Shows ensemble accuracy and comparison

---

### 5️⃣ **ensembleAll.py** - 3-Model Ensemble
**What it does:**
- Combines predictions from all three models
- Maximum robustness and confidence
- Best for critical medical applications
- Reduces variance in predictions

**Process:**
1. Loads all three models
2. Gets predictions from each
3. Applies soft voting (averages all three)
4. Generates comprehensive metrics

**Output:**
- Shows all three models' predictions
- Displays ensemble accuracy
- Confusion matrix for ensemble
